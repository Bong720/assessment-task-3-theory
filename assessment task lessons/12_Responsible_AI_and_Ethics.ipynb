## Concept 1: The Robodebt Case Study

> Between 2016 and 2019, the Australian government used an automated system called **Robodebt** to send debt notices to welfare recipients.

### The Algorithm's Fatal Flaw

```
ATO Annual Income ÷ 26 = "Assumed" fortnightly income
```

The system **assumed** everyone earned the same amount every fortnight throughout the year.

**The problem:** Casual workers, students, and seasonal workers have variable income — they earn a lot in some weeks and nothing in others.

### The Real Harm
- ~450,000 **incorrect** debt notices were sent
- People were bullied into paying money they didn't owe
- Multiple suicides were linked to the distress
- In 2021, a $1.8 billion class-action settlement was reached
- The Federal Court declared the scheme **unlawful**

### Three Things That Should Have Happened
1. **Domain expert review** — income tax specialists would have caught the averaging flaw
2. **Human approval** for each debt notice before sending
3. **Proper testing** on edge cases (casual workers, seasonal workers)

````python
# 👇 RUN THIS CELL 👇
# Demonstrate the Robodebt averaging flaw with code

# A casual worker's actual earnings across 26 fortnights
actual_earnings = np.array([
    0, 0, 1200, 1400, 0, 0, 1300,       # Jan–Apr
    1800, 2000, 1900, 0, 0, 0,           # May–Jul (busy)
    600, 700, 500, 0, 0, 0, 0,           # Aug–Oct
    0, 0, 0, 0, 0, 0                      # off-season
])

annual_total     = actual_earnings.sum()
robodebt_assumed = annual_total / 26   # the flawed calculation

print("=== Robodebt Averaging Flaw ===")
print(f"Annual income (reported to ATO): ${annual_total:,.0f}")
print(f"Robodebt assumed per fortnight:  ${robodebt_assumed:,.2f}")
print()

welfare_fortnights = [i for i, e in enumerate(actual_earnings) if e == 0]
print(f"Fortnights where worker earned $0 (and claimed welfare): {welfare_fortnights}")
print(f"Robodebt said they earned ${robodebt_assumed:.2f} during those fortnights.")
print(f"Robodebt said they owed money. They actually owed NOTHING.")

plt.figure(figsize=(10, 3))
plt.bar(range(26), actual_earnings, color='steelblue', alpha=0.7, label='Actual earnings')
plt.axhline(robodebt_assumed, color='red', linestyle='--', linewidth=2,
            label=f'Robodebt assumed (${robodebt_assumed:.0f})')
plt.xlabel('Fortnight number'); plt.ylabel('Income ($)')
plt.title('The Flaw: Assuming the Same Income Every Fortnight')
plt.legend(); plt.tight_layout(); plt.show()
````

## Concept 2: Types of Algorithmic Bias

| Bias Type | What it means | Example |
|---|---|---|
| **Historical** | Training data reflects past discrimination | Loan model trained when women were unfairly denied loans |
| **Selection** | Training data doesn't include everyone | Face recognition trained only on one demographic |
| **Measurement** | The wrong feature is used as a proxy | Using postcode to predict creditworthiness |
| **Automation** | Humans trust the AI without checking | Doctor accepts wrong AI diagnosis without question |

### Detecting Bias — The 80% Rule
One way to measure bias: **Disparate Impact Ratio**

> P(positive outcome for Group A) / P(positive outcome for Group B)

If this ratio is **below 0.8**, there may be discrimination against Group B (the 80% Rule).

````python
# 👇 RUN THIS CELL 👇
# Measure disparate impact in a simulated loan dataset

np.random.seed(42)
n = 300
group = np.random.choice(['Group A','Group B'], n)

# Group B is approved at a lower rate — systemic bias baked in
approvals = []
for g in group:
    if g == 'Group A':
        approvals.append(np.random.choice([0, 1], p=[0.30, 0.70]))   # 70% approved
    else:
        approvals.append(np.random.choice([0, 1], p=[0.45, 0.55]))   # 55% approved

df_loans = pd.DataFrame({'group': group, 'approved': approvals})

rate_a = df_loans[df_loans['group']=='Group A']['approved'].mean()
rate_b = df_loans[df_loans['group']=='Group B']['approved'].mean()
di     = rate_b / rate_a

print(f"Group A approval rate: {rate_a:.1%}")
print(f"Group B approval rate: {rate_b:.1%}")
print(f"\nDisparate Impact Ratio: {di:.3f}")
if di < 0.8:
    print("⚠️  Below 0.8 — evidence of potential discrimination against Group B")
else:
    print("✅  Above 0.8 — no significant disparate impact detected")
````

## Concept 3: An Ethical Checklist for AI Systems

Before deploying any AI that affects real people, ask these 5 questions:

| ✅ Check | Question |
|---|---|
| **Fairness** | Does it treat all groups equitably? |
| **Transparency** | Can we explain WHY it makes each decision? |
| **Accountability** | Who is responsible if it makes a mistake? |
| **Privacy** | Does it collect only the data it needs? |
| **Beneficence** | Does it do more good than harm? |

If any answer is "No" or "We don't know", the system needs more work before deployment.

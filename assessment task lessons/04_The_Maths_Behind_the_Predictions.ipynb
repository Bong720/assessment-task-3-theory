## Concept 1: The Formula — y = wx + b

When the AI learns from data, it finds the **best straight line** through the points.

That line is described by this formula:

> **ŷ = w × x + b**

| Symbol | Name | Meaning |
|---|---|---|
| **ŷ** | Prediction | The score the AI predicts |
| **x** | Input (feature) | Hours studied |
| **w** | Weight (slope) | How much y changes per 1 hour |
| **b** | Bias (intercept) | Predicted score if you study 0 hours |

When you call `.fit()`, the AI is finding the best `w` and `b` to make the line match the data.

````python
# 👇 RUN THIS CELL 👇

# Train a model and then READ what it learned
X = np.array([[1],[2],[3],[4],[5],[6],[7],[8]])
y = np.array([42, 49, 55, 60, 66, 71, 78, 83])

model = LinearRegression()
model.fit(X, y)

# .coef_ gives us w (the slope)
# .intercept_ gives us b (the intercept)
w = model.coef_[0]
b = model.intercept_

print(f"✅ AI has trained! Here is what it learned:")
print(f"   w (slope)     = {w:.2f}")
print(f"   b (intercept) = {b:.2f}")
print(f"\n   Full equation: ŷ = {w:.2f} × hours + {b:.2f}")
print(f"\n   Interpretation: Every 1 extra hour of study")
print(f"   adds about {w:.1f} marks to the predicted score.")
````

````python
# 👇 RUN THIS CELL 👇
# Visualise the line the AI learned

x_line = np.linspace(0, 9, 100)
y_line = w * x_line + b

plt.figure(figsize=(7, 4))
plt.scatter(X, y, color='steelblue', s=60, zorder=5, label='Real data')
plt.plot(x_line, y_line, color='red', linewidth=2, label=f'ŷ = {w:.1f}x + {b:.1f}')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('The Line the AI Learned')
plt.legend()
plt.tight_layout()
plt.show()
````

---
## Concept 2: Cost — How Wrong is the AI?

After the AI draws a line, we need to measure **how wrong it is**.

For each data point, the **error (residual)** is:
> **error = actual − predicted**

We want small errors. We use **RMSE** (Root Mean Squared Error) as our measure.
**Lower RMSE = Better model** ✅

````python
# 👇 RUN THIS CELL 👇
# Visualise the errors (residuals) — the vertical red lines

y_pred = model.predict(X)
errors = y - y_pred

plt.figure(figsize=(7, 4))
plt.scatter(X, y, color='steelblue', s=60, zorder=5, label='Actual')
plt.plot(x_line, y_line, color='red', linewidth=2, label='Prediction line')

# Draw error lines
for xi, yi, ypi in zip(X.ravel(), y, y_pred):
    plt.plot([xi, xi], [yi, ypi], 'k--', linewidth=1.5, alpha=0.6)

plt.xlabel('Hours Studied')
plt.ylabel('Score')
plt.title('Errors = gap between actual and predicted')
plt.legend()
plt.tight_layout()
plt.show()

rmse = np.sqrt(mean_squared_error(y, y_pred))
print(f"RMSE = {rmse:.2f}  ← average error in marks")
````

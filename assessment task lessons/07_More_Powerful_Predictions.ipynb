## Concept 1: Multiple Features

In Lesson 01 we used `[[hours]]` — one column.
For multiple features, we add more columns:

```python
# One feature:   [[2], [4], [6]]
# Two features:  [[2, 72], [4, 85], [6, 91]]
#                  ↑ hours  ↑ assignment_avg
```

The AI still uses the same `.fit()` and `.predict()` commands. The only difference is the shape of X.

````python
# 👇 RUN THIS CELL 👇

# Student data: [hours_studied, assignment_avg] → final exam mark
X_students = np.array([
    [2, 55],   # 2 hrs, assignment avg 55 → mark 47
    [3, 60],   # 3 hrs, assignment avg 60 → mark 54
    [5, 70],   # etc.
    [6, 75],
    [7, 80],
    [8, 82],
    [9, 90],
    [10, 95],
])
y_marks = np.array([47, 54, 64, 70, 77, 81, 89, 95])

model = LinearRegression()
model.fit(X_students, y_marks)

print("✅ Multi-feature model trained!")
print(f"\nWeights learned:")
print(f"  w for hours_studied:  {model.coef_[0]:.3f}")
print(f"  w for assignment_avg: {model.coef_[1]:.3f}")
print(f"  b (intercept):        {model.intercept_:.3f}")
print(f"\n🤖 Prediction for 6 hours, assignment avg 78:")
print(f"   {model.predict([[6, 78]])[0]:.1f} marks")
````

---
## Concept 2: Feature Scaling

There is a problem when features have very different ranges:
- `hours_studied`: ranges 1–10 (small numbers)
- `assignment_avg`: ranges 50–100 (bigger numbers)

The AI can struggle because the gradients for different features are very different sizes.

**Solution: StandardScaler** — rescale every feature to have mean=0 and std=1.

> Think of it like converting all measurements to the same unit.

````python
# 👇 RUN THIS CELL 👇

X_tr, X_te, y_tr, y_te = train_test_split(
    X_students, y_marks, test_size=0.25, random_state=0
)

# Scale: fit on training data, apply same scaling to test data
scaler = StandardScaler()
X_tr_scaled = scaler.fit_transform(X_tr)    # learn mean/std AND transform
X_te_scaled = scaler.transform(X_te)        # use SAME mean/std (don't re-learn!)

print("Before scaling — first training row:", X_tr[0])
print("After scaling  — first training row:", X_tr_scaled[0].round(3))

# Train on scaled data
scaled_model = LinearRegression()
scaled_model.fit(X_tr_scaled, y_tr)

y_pred = scaled_model.predict(X_te_scaled)
rmse   = np.sqrt(mean_squared_error(y_te, y_pred))
print(f"\n📊 Scaled model RMSE: {rmse:.2f} marks")

# Predict for a new student (MUST scale the new input too!)
new_student = scaler.transform([[7, 82]])
print(f"\n🤖 Prediction for 7 hrs, assignment 82: {scaled_model.predict(new_student)[0]:.1f}")
````

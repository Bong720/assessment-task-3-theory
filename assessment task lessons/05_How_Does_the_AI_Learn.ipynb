## Concept 1: The Intuition — Rolling a Ball Downhill

Imagine the AI is standing on a hilly landscape in thick fog. Its goal is to reach the **lowest valley**.

It can't see far, but it can feel which direction the ground slopes downward. So it takes small steps in the downhill direction.

That's gradient descent!

```
Cost ↑
      ╲          ← AI starts here
       ╲
        ╲___
            ╲_____
                  ╲___ ← minimum cost (goal!)
                       ─────────────────────→ w
```

Each step, the AI asks: *"If I increase `w` a tiny bit, does the cost go up or down?"*
Then it moves `w` in the direction that **reduces the cost**.

---
## Concept 2: The Update Rule

At each step (called an *iteration*), the AI adjusts `w` using this rule:

> **w_new = w_old − learning_rate × gradient**

| Term | Meaning |
|---|---|
| `learning_rate` | How big each step is (e.g., 0.01) |
| `gradient` | Which direction is uphill — we go the opposite way |

We repeat this hundreds or thousands of times until the cost stops decreasing. That's called **convergence**.

````python
# 👇 RUN THIS CELL 👇
# Let's see gradient descent in action — step by step

# Our data: hours → score (same as before)
X = np.array([2.0, 4.0, 6.0, 8.0])
y = np.array([50.0, 60.0, 70.0, 80.0])

# Helper functions
def predict(X, w, b):
    return w * X + b

def compute_cost(X, y, w, b):
    errors = y - predict(X, w, b)
    return np.mean(errors ** 2)   # Mean Squared Error

def compute_gradient(X, y, w, b):
    errors = predict(X, w, b) - y   # predicted minus actual
    dw = (2 / len(X)) * np.dot(X, errors)
    db = (2 / len(X)) * np.sum(errors)
    return dw, db

# Run gradient descent
w, b = 0.0, 0.0            # start with a bad guess
learning_rate = 0.005
cost_history = []

print(f"{'Iteration':>12} {'w':>10} {'b':>10} {'Cost':>12}")
for i in range(1001):
    cost = compute_cost(X, y, w, b)
    cost_history.append(cost)
    dw, db = compute_gradient(X, y, w, b)
    w = w - learning_rate * dw
    b = b - learning_rate * db
    if i % 200 == 0:
        print(f"{i:>12} {w:>10.4f} {b:>10.4f} {cost:>12.4f}")

print(f"\n✅ Final: w = {w:.4f}, b = {b:.4f}")
print(f"   (sklearn would get: w ≈ 5.0, b ≈ 40.0)")
````

````python
# 👇 RUN THIS CELL 👇
# Visualise the cost dropping over time

plt.figure(figsize=(7, 3))
plt.plot(cost_history, color='steelblue', linewidth=2)
plt.xlabel('Iteration')
plt.ylabel('Cost (MSE)')
plt.title('Cost Decreasing Over Time — This is Learning!')
plt.tight_layout()
plt.show()
````

## Concept 3: The Learning Rate — Goldilocks Problem

The learning rate controls **how big each step is**. Choosing it is tricky:

| Learning Rate | What happens |
|---|---|
| **Too small** (e.g., 0.00001) | Takes forever to converge — too many tiny steps |
| **Just right** (e.g., 0.005) | Converges smoothly and quickly ✅ |
| **Too large** (e.g., 0.5) | Overshoots the minimum — bounces around or explodes! |

````python
# 👇 RUN THIS CELL 👇
# Compare three learning rates

X_lr = np.array([2.0, 4.0, 6.0, 8.0])
y_lr = np.array([50.0, 60.0, 70.0, 80.0])

fig, axes = plt.subplots(1, 3, figsize=(13, 3))
rates  = [0.0001, 0.005, 0.08]
labels = ["Too Small (0.0001)", "Just Right (0.005)", "Too Large (0.08)"]
colours = ['orange', 'green', 'red']

for ax, lr, label, colour in zip(axes, rates, labels, colours):
    w_t, b_t = 0.0, 0.0
    costs_t = []
    try:
        for _ in range(300):
            c = compute_cost(X_lr, y_lr, w_t, b_t)
            if not np.isfinite(c) or c > 1e6: break
            costs_t.append(c)
            dw, db = compute_gradient(X_lr, y_lr, w_t, b_t)
            w_t -= lr * dw; b_t -= lr * db
        ax.plot(costs_t, color=colour, linewidth=2)
    except:
        ax.text(0.3, 0.5, 'DIVERGED!', transform=ax.transAxes, color='red', fontsize=12)
    ax.set_title(label, fontsize=9)
    ax.set_xlabel('Iterations'); ax.set_ylabel('Cost')

plt.tight_layout(); plt.show()
````

### Exercise 5 — Sample answer:
```python
def gradient_descent(X, y, learning_rate, iterations):
    w, b = 0.0, 0.0
    costs = []
    n = len(X)
    for _ in range(iterations):
        costs.append(np.mean((y - (w*X+b))**2))
        err = (w*X+b) - y
        w -= learning_rate * (2/n) * np.dot(X, err)
        b -= learning_rate * (2/n) * np.sum(err)
    return w, b, costs

w_f, b_f, _ = gradient_descent(X_ins, y_ins, 0.001, 2000)
rmse = np.sqrt(np.mean((y_ins - (w_f*X_ins+b_f))**2))
print(f"wingsize = {w_f:.2f} × latitude + {b_f:.2f}  |  RMSE={rmse:.2f}")
```

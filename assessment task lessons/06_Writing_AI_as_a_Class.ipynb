## Concept 1: What is a Class?

A **Class** is a blueprint. An **Object** is a copy made from that blueprint.

Think of it like a cookie cutter (class) vs actual cookies (objects):

```python
# The blueprint
class Cookie:
    def __init__(self, flavour):
        self.flavour = flavour      # attribute

    def describe(self):             # method
        print(f"I am a {self.flavour} cookie!")

# Making objects from the blueprint
cookie1 = Cookie("chocolate")
cookie2 = Cookie("vanilla")
cookie1.describe()   # → "I am a chocolate cookie!"
cookie2.describe()   # → "I am a vanilla cookie!"
```

| OOP Term | Meaning | Our AI Example |
|---|---|---|
| **Class** | The blueprint | `MyLinearRegression` |
| **Object** | A copy of the blueprint | `model = MyLinearRegression()` |
| **`__init__`** | Runs when the object is created | Sets starting values for `w` and `b` |
| **Attribute** | Data stored in the object | `model.w`, `model.b` |
| **Method** | A function inside the class | `model.fit()`, `model.predict()` |
| **`self`** | Refers to "this object" | Used in every method |

## Concept 2: Build the Class — One Method at a Time

We'll build `MyLinearRegression` step by step, adding one method at a time.

````python
# 👇 RUN THIS CELL 👇
# Step 1: The constructor — what happens when you create a new model?

class MyLinearRegression:

    def __init__(self, learning_rate=0.005, iterations=1000):
        """
        Called when you write: model = MyLinearRegression()
        Sets up starting values.
        """
        self.learning_rate = learning_rate
        self.iterations    = iterations
        self.w = 0.0    # slope — starts at 0, will be learned
        self.b = 0.0    # intercept — starts at 0, will be learned

# Test it
model = MyLinearRegression(learning_rate=0.01)
print(f"✅ New model created!")
print(f"   w = {model.w}  (will be learned during training)")
print(f"   b = {model.b}  (will be learned during training)")
print(f"   learning_rate = {model.learning_rate}")
````

````python
# 👇 RUN THIS CELL 👇
# Step 2: Add the fit() method — gradient descent lives here

class MyLinearRegression:

    def __init__(self, learning_rate=0.005, iterations=1000):
        self.learning_rate = learning_rate
        self.iterations    = iterations
        self.w = 0.0
        self.b = 0.0
        self.cost_history  = []

    def fit(self, X, y):
        """
        Train the model: find the best w and b using gradient descent.
        X must be a 1D array of inputs.
        y must be a 1D array of correct answers.
        """
        n = len(X)
        self.cost_history = []

        for _ in range(self.iterations):
            y_pred = self.w * X + self.b
            error  = y_pred - y

            dw = (2 / n) * np.dot(X, error)
            db = (2 / n) * np.sum(error)

            self.w -= self.learning_rate * dw
            self.b -= self.learning_rate * db

            self.cost_history.append(np.mean(error ** 2))

        return self   # allows chaining

# Test it
X_train = np.array([2.0, 4.0, 6.0, 8.0])
y_train = np.array([50.0, 60.0, 70.0, 80.0])

model = MyLinearRegression(learning_rate=0.005, iterations=1000)
model.fit(X_train, y_train)
print(f"✅ Training done!")
print(f"   w = {model.w:.4f}  (sklearn gets ~5.0)")
print(f"   b = {model.b:.4f}  (sklearn gets ~40.0)")
````

````python
# 👇 RUN THIS CELL 👇
# Step 3: Add predict() and evaluate() methods

class MyLinearRegression:

    def __init__(self, learning_rate=0.005, iterations=1000):
        self.learning_rate = learning_rate
        self.iterations    = iterations
        self.w = 0.0; self.b = 0.0
        self.cost_history  = []

    def fit(self, X, y):
        n = len(X)
        self.cost_history = []
        for _ in range(self.iterations):
            error = (self.w * X + self.b) - y
            self.w -= self.learning_rate * (2/n) * np.dot(X, error)
            self.b -= self.learning_rate * (2/n) * np.sum(error)
            self.cost_history.append(np.mean(error**2))
        return self

    def predict(self, X):
        """Use the learned w and b to make predictions."""
        return self.w * X + self.b

    def evaluate(self, X, y):
        """Print RMSE for this model."""
        rmse = np.sqrt(np.mean((y - self.predict(X))**2))
        print(f"📊 RMSE = {rmse:.4f}")
        return rmse

# Full demonstration
model = MyLinearRegression(learning_rate=0.005, iterations=1000)
model.fit(X_train, y_train)

print("Predictions:")
for h in [3, 5, 7, 9]:
    print(f"  {h} hours → {model.predict(h):.1f} marks")

print()
model.evaluate(X_train, y_train)
print(f"\n🤖 Predicting for 5 hours: {model.predict(5):.1f} marks")
````

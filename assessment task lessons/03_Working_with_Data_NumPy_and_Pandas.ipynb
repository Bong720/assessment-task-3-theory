## Concept 1: NumPy Arrays

NumPy (Numerical Python) gives us **arrays** — lists of numbers that can do maths super fast.

### Why not just use Python lists?
| | Python List | NumPy Array |
|---|---|---|
| Speed | Slow for maths | Very fast |
| Maths operations | Need a loop | Automatic! |
| Used by AI? | No | Yes — every AI library uses NumPy |

````python
# 👇 RUN THIS CELL 👇

# Creating a NumPy array from a list
scores = np.array([55, 72, 48, 91, 63, 80, 44, 77])

print(f"Scores:      {scores}")
print(f"Add 5:       {scores + 5}")      # adds 5 to EVERY element at once!
print(f"Times 2:     {scores * 2}")
print(f"\nMean:   {scores.mean():.1f}")
print(f"Max:    {scores.max()}")
print(f"Min:    {scores.min()}")
print(f"Sorted: {np.sort(scores)}")
````

````python
# 👇 RUN THIS CELL 👇
# Boolean filtering — super useful for finding patterns!

scores = np.array([55, 72, 48, 91, 63, 80, 44, 77])

# Find all scores above 70
passing = scores[scores >= 70]

print(f"All scores:     {scores}")
print(f"Passing (≥70):  {passing}")
print(f"Number passing: {len(passing)} out of {len(scores)}")
print(f"Pass rate:      {len(passing)/len(scores)*100:.0f}%")
````

## Concept 2: Pandas DataFrames

A **DataFrame** is a table — rows and columns, just like a spreadsheet.

Pandas lets us load CSV files, filter rows, and explore data easily.

**Key commands:**
* `pd.read_csv('file.csv')` — load a CSV file
* `df.head()` — see the first 5 rows
* `df['column']` — select one column
* `df[df['column'] > value]` — filter rows

````python
# 👇 RUN THIS CELL 👇

# Create a DataFrame from scratch (like a mini spreadsheet)
data = {
    'student': ['Alice', 'Bob', 'Charlie', 'Diana', 'Ethan'],
    'hours':   [3.5, 6.0, 2.0, 7.5, 5.0],
    'mark':    [65,  82,  51,  91,  74]
}
df = pd.DataFrame(data)

print("=== Student DataFrame ===")
print(df)
print(f"\nShape: {df.shape}  ({df.shape[0]} rows, {df.shape[1]} columns)")
````

````python
# 👇 RUN THIS CELL 👇
# Load the REAL insects dataset that came with this course!

df_bugs = pd.read_csv('insects.csv', sep='\t')  # sep='\t' means tab-separated

print("=== Insects Dataset (first 5 rows) ===")
print(df_bugs.head())
print(f"\nTotal rows: {len(df_bugs)}")
print(f"Columns: {list(df_bugs.columns)}")
print(f"\nColumn data types:")
print(df_bugs.dtypes)
````

````python
# 👇 RUN THIS CELL 👇
# Explore and filter the data

print("=== Basic Statistics ===")
print(df_bugs.describe().round(1))

print("\n=== Filter: Bugs with wingsize > 920mm ===")
big_bugs = df_bugs[df_bugs['wingsize'] > 920]
print(big_bugs.head())
print(f"Number of big bugs: {len(big_bugs)}")

print("\n=== Average wing size by sex ===")
print(df_bugs.groupby('sex')['wingsize'].mean().round(1))
print("(0 = female, 1 = male)")
````

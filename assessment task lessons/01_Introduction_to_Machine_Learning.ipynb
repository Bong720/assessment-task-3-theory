## Concept 1: What is the difference?

How is Artificial Intelligence (AI) different from normal computer programming?

### Normal Programming (You write the rules)
In normal coding, **YOU** have to type out every single rule. The computer just follows your exact instructions.

> **Formula:** `Data + Human-Written Rules = Outputs`

**Example:** If you want a program to grade a test, you write the rule yourself:
*`if score > 50, then print "Pass"`*

---

### Machine Learning (The computer learns the rules)
But what if you want a program to look at a photo and say if it's a Dog or a Cat? You can't write a rule for that!

Instead, we use **Machine Learning**.
We give the computer thousands of examples (Answers), and ask the computer to figure out the rules mathematically.

> **Formula:** `Data + Answers = The Computer Learns the Rules`

---
## Concept 2: Let's Build an AI together!

We are going to build a small AI that can predict a student's test score based on how many hours they studied.

We will do this in 4 tiny steps.

### Step 1: Give the AI some Data
An AI needs examples to learn from. Let's give it the data of 4 past students.

* **X (Inputs):** The hours they studied (2, 4, 6, and 8 hours).
* **y (Answers):** The scores they actually got (50, 60, 70, and 80).

````python
# 👇 RUN THIS CELL 👇

# The Inputs (Notice the double brackets [[ ]] — this is how the AI likes to read inputs)
X_hours = np.array([[2], [4], [6], [8]])

# The Answers
y_scores = np.array([50, 60, 70, 80])

print("✅ Data has been saved into the computer's memory!")
````

### Step 2: Create the "Brain"
Now we need to create an empty AI model. Think of this as a newborn brain — it doesn't know anything yet.

We will use a tool called `LinearRegression()`. This is a math tool that looks for straight-line patterns.

````python
# 👇 RUN THIS CELL 👇

# Create the empty AI and call it 'my_ai'
my_ai = LinearRegression()

print("✅ Empty AI created! Right now, it knows nothing.")
````

### Step 3: Train the Brain!
This is where the magic happens. We use a command called `.fit()`.

`.fit()` tells the AI: *"Look at the inputs, look at the answers, and find the hidden pattern!"*

````python
# 👇 RUN THIS CELL 👇

# Train the AI using our data from Step 1
my_ai.fit(X_hours, y_scores)

print("✅ Training complete! The AI has found the pattern.")
````

### Step 4: Make a Prediction
Now that the AI is trained, we can ask it a new question.
Let's ask it: *"What score will someone get if they study for 5 hours?"*

We use the `.predict()` command for this.

````python
# 👇 RUN THIS CELL 👇

# Ask the AI to predict the score for 5 hours.
# (Again, we use double brackets [[ ]] for the input)
prediction = my_ai.predict([[5]])

print(f"🤖 The AI predicts that 5 hours of study will result in a score of: {prediction[0]:.0f}")
````

## Concept 3: The Three Flavours of AI

You just built a **Supervised Learning** model! But there are actually 3 different ways to teach an AI:

### 1. Supervised Learning (Learning with a Teacher)
* **How it works:** You give the AI the data AND the correct answers (just like we did with the test scores).
* **Used for:** Predicting numbers (like house prices) or categorising things (like Spam vs Not Spam).

### 2. Unsupervised Learning (Learning by Exploring)
* **How it works:** You give the AI a giant pile of data, but NO answers. You ask it to find hidden groups on its own.
* **Used for:** Spotify looking at millions of users and grouping similar songs into a playlist.

### 3. Reinforcement Learning (Learning by Trial and Error)
* **How it works:** The AI plays a "game". It gets a point for doing something good, and loses a point for doing something bad.
* **Used for:** Teaching a robot to walk, or teaching a self-driving car to stay on the road.
